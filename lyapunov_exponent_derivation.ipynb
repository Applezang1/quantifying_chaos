# Derivation of the Lyapunov Exponent from the Divergence Equation

Define $x_0 $ as the initial point and define $x_0 + S_0$ as a nearby point separated by a distance of $S_0$

With this setup, define the following expression

$$|S_n| = |S_0| \, e^{n\lambda}$$
Where: 
- n represents the timestep
- $S_n$ represents the distance between the two points at that specific timestep

This represents the Lyapunov divergence equation because of the presence of an exponential term ($e^{n\lambda}$), which states that the distance between the two points exponentially converge or diverge over timesteps. 

The Lyapunov exponent is defined as $\lambda$, where $\lambda$ > 0 means that the distance between the two points are exponentially diverging over time and $\lambda$ < 0 means that the distance between the two points are exponentially converging. 

By taking the natural log of both sides and expanding the expression, we are able to write the Lyapunov divergence equation in point-slope form

$$\ln S_n = \lambda n + \ln S_0$$

We are able to now determine the Lyapunov exponent by plotting the natural log of the distance between the two points over timesteps and computing the slope of that graph.
