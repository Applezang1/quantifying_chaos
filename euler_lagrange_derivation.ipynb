# Euler-Lagrange Equation Derivation

<img src="setup.png" style="display:block; margin-left:auto; margin-right:auto;" width="400">

We imagine an arbitrary function, η(x) connecting two points. This function is random, and can be any of all possible curves that can connect the two points. This means its only restriction is that it must have zero variation at x_1 and x_2. y(x) represents the optimal function to connect the two points. In the case of the double pendulum, it represents the path of least action. Epsilon (ε) is an amount that η(x) is scaled by. By multiplying the two, we can account for all possible curves that connect the two points.

Another arbitrary function is ȳ. It will represent the “testing” function we will use to try to find the optimal function. Notice that as εη(x) approaches zero, ȳ(x) approaches y(x). 

$$
\tilde{y}(x) = y(x) + \varepsilon\, \eta(x) \tag{1}
$$

We differentiate the equation and get 

$$
\tilde{y}'(x) = y'(x) + \varepsilon\, \eta'(x) \tag{2}
$$

Our integral is the functional of x, y(x) and y’(x). In the case of the double pendulum, these functions will be able to describe the system completely as mathematical terms. We can obtain the specific functions for these equations by finding the Lagrangian of the system.

$$
I = \int_{x_1}^{x_2} F\big(x, y(x), y'(x)\big)\, dx \tag{3}
$$

The optimal path exists when the variation of η(x) is zero, meaning it is an extremal. Just like in basic calculus, we can take the derivative and set it to zero. It is a whole derivative and not a partial derivative because ε is the only value that varies. As a reminder, as ε approaches zero, ȳ will become more like y, our optimal path. 

$$
\left. \frac{dI}{d\varepsilon} \right|_{\varepsilon = 0}
= \left. \frac{d}{d\varepsilon} \int_{x_1}^{x_2}
F\big(x, \tilde{y}(x), \tilde{y}'(x)\big)\, dx \right|_{\varepsilon=0}
= 0 \tag{4}
$$


By Leibniz's rule:

$$
\int_{x_1}^{x_2}
\left. \frac{d}{d\varepsilon} F\big(x, \tilde{y}(x), \tilde{y}'(x)\big)
\right|_{\varepsilon=0} dx = 0 \tag{5}
$$

The whole derivative of the functional can be expressed as partial derivatives of each of its components. Notice that the first term does not vary with ε, so it cancels to zero. 

$$
\int_{x_1}^{x_2}
\left( 
\frac{\partial F}{\partial y}\frac{\partial \tilde{y}}{\partial \varepsilon}
+
\frac{\partial F}{\partial y'}\frac{\partial \tilde{y}'}{\partial \varepsilon}
\right)_{\varepsilon=0}
dx = 0 \tag{6}
$$

Taking the derivative of equation (1) and (2) with respect to ε gives

$$
\frac{\partial \tilde{y}}{\partial \varepsilon} = \eta(x) \tag{7}
$$


and

$$
\frac{\partial \tilde{y}'}{\partial \varepsilon} = \eta'(x) \tag{8}
$$

By substituting Equations (7) and (8) in our integral, we get the weak form of the Euler-Lagrange equation. 

$$
\int_{x_1}^{x_2}
\left(
\frac{\partial F}{\partial y}\,\eta(x)
+
\frac{\partial F}{\partial y'}\,\eta'(x)
\right)
dx = 0 \tag{9}
$$


If we split up the integral and integrate the second term by parts we get

$$
\int_{x_1}^{x_2} 
\frac{\partial F}{\partial y}\,\eta
+
\left. \frac{\partial F}{\partial y'} \eta \right|_{x_1}^{x_2}
-
\int_{x_1}^{x_2} 
\frac{d}{dx}\left( \frac{\partial F}{\partial y'} \right)\eta \, dx
= 0
\tag{10}
$$


Once again, η is arbitrary and predetermined, so the first term will be zero. 

$$
\int_{x_1}^{x_2}
\left(
\frac{\partial F}{\partial y}
-
\frac{d}{dx}\left( \frac{\partial F}{\partial y'} \right)
\right)\eta(x)\, dx
= 0
\tag{11}
$$

Because η is arbitrary, the other term in the integral must be equal to zero for the full equation to be zero. 

$$
\frac{\partial F}{\partial y}
-
\frac{d}{dx}\left( \frac{\partial F}{\partial y'} \right)
= 0
\tag{12}
$$

This yields the complete Euler-Lagrange Equation.
